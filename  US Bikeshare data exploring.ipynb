```python
import time
import pandas as pd
import numpy as np
```

```python
CITY_DATA = { 'Chicago': 'chicago.csv',
              'New York City': 'new_york_city.csv',
              'Washington': 'washington.csv' }
```

```python
def get_filters():
    """
    Asks user to specify a city, month, and day to analyze.

    Returns:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    """
    print('Hello! Let\'s explore some US bikeshare data!')
    # TO DO: get user input for city (chicago, new york city, washington). HINT: Use a while loop to handle invalid inputs
    while True:
      city=input('enter your city name: ').title()
      if city.lower() not in ('chicago', 'new york city', 'washington'):# CITY_DATA.keys():
        print('sorry, please enter a correct city name')
        continue
      else:
        print('you have choosen {} city'.format(city))
        break


    # TO DO: get user input for month (all, january, february, ... , june)
    while True:
      month=input('enter a valid month name: ').title()
      if (month.lower()) in('january','february','march','april','may','june','july','august','sept','october','november','december'):
          print('you have choosen {}'.format(month.lower()))
          break
      elif(month.lower())=='all':
          print('you have choosen all months!')
          break
      else:
          print("invalid date, please check your spelling and enter correct month name!")
          month=input('enter a month day name: ').title()
          

    # TO DO: get user input for day of week (all, monday, tuesday, ... sunday)
    while True:
        day=input('enter a valid day name: ').title()
        if (day.lower()) in('monday','tuesday','wednesday','thursday','friday','saturday','sunday'):
          print('you have choosen {}'.format(day.lower()))
          break
        elif(day.lower())=='all':
          print('you have choosen all days!')
          break
        else:
          print("invalid date, please check your spelling and enter correct day name!")
          day=input('enter a valid day name: ').title()
          
    print('now you have choosen '+city+' city, during '+month+' month and '+day+' day!')
    print('-'*40)
    return city, month, day


def load_data(city, month, day):
    """
    Loads data for the specified city and filters by month and day if applicable.

    Args:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    Returns:
        df - Pandas DataFrame containing city data filtered by month and day
    """
    df = pd.read_csv(CITY_DATA[city])

    return df


def time_stats(df):
    """Displays statistics on the most frequent times of travel."""

    print('\nCalculating The Most Frequent Times of Travel...\n')
    start_time = time.time()

    # TO DO: display the most common month
    df['Start Time'] = pd.to_datetime(df['Start Time'])

    df['month'] = df['Start Time'].dt.month
    print("the most common month is the " + str(df.month.mode()[0]) + " month")


    # TO DO: display the most common day of week
    df['day_of_week'] = df['Start Time'].dt.weekday
    print("the most common day of the week is the " + str(df.day_of_week.mode()[0]) +" day of the week")


    # TO DO: display the most common start hour
    df['hour'] = df['Start Time'].dt.hour
    df.hour.mode()[0]
    print("the most common hour of the day is " + str(df.hour.mode()[0]) +" O'clock")


    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def station_stats(df):
    """Displays statistics on the most popular stations and trip."""

    print('\nCalculating The Most Popular Stations and Trip...\n')
    start_time = time.time()

    # TO DO: display most commonly used start station
    print("the most common starting stations is " +df['Start Station'].mode()[0])

    # TO DO: display most commonly used end station
    print("the most common starting stations is " +df['End Station'].mode()[0])


    # TO DO: display most frequent combination of start station and end station trip
    print("the most common combination stations stations are "+ (df['Start Station']+' AND '+df['End Station']).mode()[0])


    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def trip_duration_stats(df):
    """Displays statistics on the total and average trip duration."""

    print('\nCalculating Trip Duration...\n')
    start_time = time.time()

    # TO DO: display total travel time
    print("total travel time is " + str(df['Trip Duration'].sum()) + " seconds"+ ".. which almost " + str(round(df['Trip Duration'].sum()/60/60)) + " hours")


    # TO DO: display mean travel time
    print("total travel time is " + str(df['Trip Duration'].mean()) + " seconds"+ ".. which almost " + str(round(df['Trip Duration'].mean()/60)) + " minutes")


    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def user_stats(df):
    """Displays statistics on bikeshare users."""

    print('\nCalculating User Stats...\n')
    start_time = time.time()

    # TO DO: Display counts of user types
    print("number of users based on user type is the following: \n")
    print(df['User Type'].groupby(df['User Type']).count())
    print("\n")

    # TO DO: Display counts of gender
    print("number of users based on gender is the following: \n")
    try:
      print(df['Gender'].groupby(df['Gender']).count())
      print("\n \n ............")
    except:
      print("this city doesn't have a gender field on it.\n we have to skip this point!")

    # TO DO: Display earliest, most recent, and most common year of birth
    try:
      youngest=round(df['Birth Year'].max())
      oldest=round(df['Birth Year'].min())

      print("the youngest user was born in "+ str(youngest)+" thus they are only " +str(2020-youngest) + " years old! ")
      print("the oldest user was born in " + str(oldest) +" thus they are only " +str(2020-oldest) + " years old! ")

      if oldest>100:
        print("this is odd, someone at the age of "+str(2020-oldest) +" ride a bike!, something must be wrong here! ")
      else:
        pass
    except:
      print("this city doesn't have a birth year field on it.\n we have to ship this point!" )

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def main():
    while True:
        city, month, day = get_filters()
        df = load_data(city, month, day)

        time_stats(df)
        station_stats(df)
        trip_duration_stats(df)
        user_stats(df)

        restart = input('\nWould you like to restart? Enter yes or no.\n')
        if restart.lower() != 'yes':
            break


if __name__ == "__main__":
    main()
```
